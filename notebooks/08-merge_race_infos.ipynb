# 08 - Merge race_infos

This notebook merges every CSV in `../data/raw/race_info` into one consolidated file.

The merge is written in a streaming way: each CSV is read and appended directly to disk, which is much more memory efficient than concatenating everything in RAM first.

If this becomes a recurring pipeline step, the same function can be moved into a `.py` module or script later.

```python
from pathlib import Path
from time import perf_counter

import pandas as pd
from IPython.display import display
from tqdm.auto import tqdm

from src.dataset.load import merge_csv_folder

DATA_DIR = Path("../data/raw")
INTER_DIR = Path("../data/intermediate")
RACE_HEADER_DIR = DATA_DIR / "race_info"
OUTPUT_PATH = INTER_DIR / "08_all_race_infos.csv"
ADD_SOURCE_FILE = True

pd.set_option("display.max_columns", 50)
pd.set_option("display.max_rows", 200)
```

```python
merge_stats, schema_mismatches, output_columns = merge_csv_folder(
    input_dir=RACE_HEADER_DIR,
    output_path=OUTPUT_PATH,
    add_source_file=ADD_SOURCE_FILE,
)

display(merge_stats.to_frame(name="value"))
```
