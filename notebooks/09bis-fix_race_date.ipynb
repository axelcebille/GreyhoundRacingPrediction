# Fix the removed `raceDate` column from the race header

```python
race_headers = pd.read_csv("../data/processed/09_all_race_header.csv")
race_headers_a_class = pd.read_csv("../data/processed/09_graded_a_race_header.csv")
race_header_unprocessed = pd.read_csv("../data/intermediate/07_all_race_header.csv")
```

```python
# Assign raceDate from the unprocessed race header to the graded A class race header
#race_headers_a_class = race_headers_a_class.drop(columns=["raceDate"], errors="ignore")
race_headers_a_class = race_headers_a_class.merge(
    race_header_unprocessed[["source_file", "raceDate"]],
    on="source_file",
    how="left",
)

# Optional: verify the assignment
race_headers_a_class["raceDate"].isna().sum(), race_headers_a_class.shape
```

```python
# Assign raceDate from the unprocessed race header to the graded A class race header
#race_headers_a_class = race_headers_a_class.drop(columns=["raceDate"], errors="ignore")
race_headers = race_headers.merge(
    race_header_unprocessed[["source_file", "raceDate"]],
    on="source_file",
    how="left",
)

# Optional: verify the assignment
race_headers["raceDate"].isna().sum(), race_headers.shape
```

```python
race_headers.to_csv("../data/processed/09_all_race_header.csv", index=False)
race_headers_a_class.to_csv("../data/processed/09_graded_a_race_header.csv", index=False)
```
