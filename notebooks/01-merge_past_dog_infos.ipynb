# Merge all dog infos

Reads every per-dog CSV in `data/raw/info_dogs`, merges them into a single DataFrame, and saves it back to `data/raw/all_dog_infos.csv`.

```python
from pathlib import Path
import pandas as pd

info_dir = Path("..") / "data" / "raw" / "info_dogs"
out_path = Path("..") / "data" / "raw" / "all_dog_infos.csv"

files = sorted(info_dir.glob("*.csv"))
dfs = []
skipped_empty = 0
for f in files:
    if f.stat().st_size == 0:
        skipped_empty += 1
        continue
    try:
        df = pd.read_csv(f)
    except pd.errors.EmptyDataError:
        skipped_empty += 1
        continue
    if df.empty:
        skipped_empty += 1
        continue
    if "dogId" not in df.columns:
        # fall back to filename when dogId column is missing
        df["dogId"] = int(f.stem) if f.stem.isdigit() else f.stem
    dfs.append(df)

all_dog_infos = pd.concat(dfs, ignore_index=True) if dfs else pd.DataFrame()
out_path.parent.mkdir(parents=True, exist_ok=True)
all_dog_infos.to_csv(out_path, index=False)

print(f"Saved {len(all_dog_infos):,} rows to {out_path}")
print(f"Skipped empty files: {skipped_empty}")
all_dog_infos.head()
```
